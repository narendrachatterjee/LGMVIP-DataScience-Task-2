```python
#importing required libraries
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading the image
img = cv2.imread('C:\\Users\\Narendra\\Desktop\\testImg.jpg')
cv2.imshow("Image",img)
cv2.waitKey(0)
```

```python
#Reading the image in RGB format
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

```python
#converting the image into grayscale
gray_IMG = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_IMG)
```

```python
#inverting the image
invert_IMG = 255-gray_IMG
plt.imshow(invert_IMG)
```

```python
#Converting the image into blurred image using Gaussian Blur function
blur_IMG = cv2.GaussianBlur(invert_IMG , (21 , 21),0)
plt.imshow(blur_IMG)
```

```python
#Inverting the blurred image to get the sktech
invert_blur_IMG = 255-blur_IMG
sketch = cv2.divide(gray_IMG, invert_blur_IMG, scale = 256.00)
plt.imshow(sketch)
```

```python
img_sketch_RGB = cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB)

plt.imshow(img_sketch_RGB)
```
